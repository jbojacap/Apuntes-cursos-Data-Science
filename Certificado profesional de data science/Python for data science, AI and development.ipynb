## Objetos y clases

```python
import matplotlib.pyplot as plt

class circulo(object):
    def __init__(self,radio,color): #Constructor
        self.radio = radio;
        self.color = color;
    def inc_radio(self,incremento): #Method para incrementar el radio
        self.radio += incremento
        return(self.radio)
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((-10, -10), radius=self.radio, fc=self.color))
        plt.axis('scaled')
        plt.show()  

mi_circulo = circulo(10,"red")
print(mi_circulo.color)
print(mi_circulo.radio)
mi_circulo.inc_radio(5) #Llamar el method para incrementar y fijar incremento en 5
print(mi_circulo.radio) 
mi_circulo.drawCircle() #Dibuja el círculo, origen en -10,-10
```

### Ejercicio

Crear una clase que:
* En el constructor debe quitar todos los signos de puntuación, asumiendo que solo hay ,.!? 
* En el method freqall se hace un diccionario con las frecuencias de las palabras
* En el method freqof se debe poder introducir una palabra y retornar el número de repeticiones de esa palabra en el texto

```python
class analysedText(object):
    
    def __init__ (self, text):
        #Eliminar signos de puntiación y dejar todo en minúscula
        textedit = text.replace("!","").replace(",","").replace(".","").replace("?","").lower()
        #Tomar la operación anterior y asignarla como atributo
        self.txtedit = textedit
    
    def freqAll(self):    
        # Se llama el atributo anterior y ya que no hay signos y está todo en minúscula, se almacenan las palabras en una lista
        lista_palabras = self.txtedit.split(" ")
        # Se crea un diccionario para almacenar las palabras y su número de repeticiones
        dicci = {}
        # Al diccionario se añaden las palabras con el número de repeticiones 
        for palabra in set(lista_palabras): #Nótese que se eliminan los duplicados al convertir la lista en set, pero esto lo hace al FINAL del bucle, también notar que el diccionario no almacena duplicados            
            dicci[palabra] = lista_palabras.count(palabra) #Se usa el method count (de listas) para sumar las repeticiones y se almacena en los valores de los índices del diccionario
        return dicci
    
    def freqOf(self, palabra):
        #Traer el diccionario, el retorno del method anterior
        diccio = self.freqAll()
        if palabra in diccio: #Se busca si la palabra está en la lista
            return diccio[palabra] #Si está, se devuelve el número de repeticiones, esto devuelve el valor del índice correspondiente
        else:
            return 0 #Si no está, devuelve un cero

texto_prueba = "Lorem ipsum dolor! diam amet, consetetur Lorem magna. sed diam nonumy eirmod tempor. diam et labore? et diam magna. et diam amet."
prueba = analysedText(texto_prueba)
print(prueba.freqAll())
print(prueba.freqOf("diam"))
```

### Ejercicio

Tomar dos listas, la primera es una lista de miembros de un club (en txt), en esta lista hay miembros activos y no activos (definidos en la lista por un "yes" o un "no"), se deben eliminar nos no activos y moverlos a la otra lista de no activos (en txt también). Esto a través de una función en Python llamada "CleanFiles" que tome las dos listas como entradas.

```python
#Este código genera aleatoriamente los archivos txt, ignorar
from random import randint as rnd

memReg = 'members.txt'
exReg = 'inactive.txt'
fee =('yes','no')

def genFiles(current,old):
    with open(current,'w+') as writefile: 
        writefile.write('Membership No  Date Joined  Active  \n')
        data = "{:^13}  {:<11}  {:<6}\n"

        for rowno in range(20):
            date = str(rnd(2015,2020))+ '-' + str(rnd(1,12))+'-'+str(rnd(1,25))
            writefile.write(data.format(rnd(10000,99999),date,fee[rnd(0,1)]))


    with open(old,'w+') as writefile: 
        writefile.write('Membership No  Date Joined  Active  \n')
        data = "{:^13}  {:<11}  {:<6}\n"
        for rowno in range(3):
            date = str(rnd(2015,2020))+ '-' + str(rnd(1,12))+'-'+str(rnd(1,25))
            writefile.write(data.format(rnd(10000,99999),date,fee[1]))
```

```python
def cleanFiles(miembros,exmiembros): #Se define una función con dos entradas, la lista de miembros y la de exmiembros (txt)
    with open(miembros,'r+') as miembross: #Se abre la lista de miembros como r+, siendo la que se va a sobreescribir
        with open(exmiembros,'a+') as exmiembross: #Se abre la lista de exmiembros, siendo la que se va a extender
            #Obtener datos
            miembross.seek(0) #Se sitúa el cursor en el origen
            members = miembross.readlines() #Se leen los renglones como elementos de una lista
            #Remover encabezado
            header = members[0] #Se almacena el encabezado para ponerlo nuevamente después, siendo el primer elemento de la lista
            members.pop(0) #Se remueve el encabezado de la lista
                
            inactive = [member for member in members if ('no' in member)] #Se examina cada renglón de los miembros, si hay un "no", se almacena en la lista de inactivos(recién creada)
            miembross.seek(0) #Se sitúa el cursor en el origen
            miembross.write(header) #Se pone de nuevo el encabezado
            for member in members: #Se examina cada miembro de la lista de miembros
                if (member in inactive): #Si el miembro está en la lista de inactivos, se añade a la lista de exmiembros
                    exmiembross.write(member) #Se añade el miembro a exmiembros
                else:
                    miembross.write(member) #Si está activo, se vuelve a escribir al inicio de la lista hasta completar la lista de miembros activos (rehacer la lista)      
        miembross.truncate() #Borra todos los datos que no se hayan puesto con el primer with en la lista de miembros, evita duplicados (borrar la primera versión)
                
memReg = 'members.txt'
exReg = 'inactive.txt'
cleanFiles(memReg,exReg)

# code to help you see the files

headers = "Membership No  Date Joined  Active  \n"

with open(memReg,'r') as readFile:
    print("Active Members: \n\n")
    print(readFile.read())
    
with open(exReg,'r') as readFile:
    print("Inactive Members: \n\n")
    print(readFile.read())
genFiles(memReg,exReg)
```
